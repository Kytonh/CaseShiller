```python
#7
#五大城市高中低端与整体房产增长率excel
def 市场(城市,整体,高端,中端,低端):
    data=pd.read_csv(整体,low_memory=False)
    data1=pd.read_csv(高端,low_memory=False)
    data2=pd.read_csv(中端,low_memory=False)
    data3=pd.read_csv(低端,low_memory=False)
    a=np.array(data)
    b=np.array(data1)
    c=np.array(data2)
    d=np.array(data3)
    x=0
    y=12
    时间=['年份']
    增长率=['整体房产']
    增长率1=['高端房产']
    增长率2=['中端房产']
    增长率3=['低端房产']
    while y<len(a):
        rate=(a[y,1]-a[x,1])/a[x,1]
        rate1=(b[y,1]-b[x,1])/b[x,1]
        rate2=(c[y,1]-c[x,1])/c[x,1]
        rate3=(d[y,1]-d[x,1])/d[x,1]
        增长率.append(rate)
        增长率1.append(rate1)
        增长率2.append(rate2)
        增长率3.append(rate3)
        y+=1
        x+=1
    for x in range (12,len(a)):
        时间.append(a[x,0])
    df=pd.DataFrame([增长率],columns=(时间))
    df1=pd.DataFrame([增长率1],columns=(时间))
    df2=pd.DataFrame([增长率2],columns=(时间))
    df3=pd.DataFrame([增长率3],columns=(时间))
    writer = pd.ExcelWriter(城市+'.xlsx', engine='xlsxwriter')
    df.to_excel(writer,startrow=0,index=False)
    df1.to_excel(writer,startrow=len(df)+1,header=False,index=False)
    df2.to_excel(writer,startrow=len(df)+2,header=False,index=False)
    df3.to_excel(writer,startrow=len(df)+3,header=False,index=False)
    writer.save()


def 芝加哥市场():
    data=pd.read_csv('CHXRNSA.csv',low_memory=False)
    data1=pd.read_csv('CHXRHTNSA.csv',low_memory=False)
    data2=pd.read_csv('CHXRMTNSA.csv',low_memory=False)
    data3=pd.read_csv('CHXRLTNSA.csv',low_memory=False)
    a=np.array(data)
    b=np.array(data1)
    c=np.array(data2)
    d=np.array(data3)
    x=0
    y=12
    u=60
    v=72
    时间=['年份']
    增长率=['整体房产']
    增长率1=['高端房产']
    增长率2=['中端房产']
    增长率3=['低端房产']
    while y<len(b):
        rate=(a[v,1]-a[u,1])/a[u,1]
        rate1=(b[y,1]-b[x,1])/b[x,1]
        rate2=(c[y,1]-c[x,1])/c[x,1]
        rate3=(d[y,1]-d[x,1])/d[x,1]
        增长率.append(rate)
        增长率1.append(rate1)
        增长率2.append(rate2)
        增长率3.append(rate3)
        y+=1
        x+=1
        u+=1
        v+=1
    for x in range (12,len(b)):
        时间.append(b[x,0])
    df=pd.DataFrame([增长率],columns=(时间))
    df1=pd.DataFrame([增长率1],columns=(时间))
    df2=pd.DataFrame([增长率2],columns=(时间))
    df3=pd.DataFrame([增长率3],columns=(时间))
    writer = pd.ExcelWriter('芝加哥.xlsx', engine='xlsxwriter')
    df.to_excel(writer,startrow=0,index=False)
    df1.to_excel(writer,startrow=len(df)+1,header=False,index=False)
    df2.to_excel(writer,startrow=len(df)+2,header=False,index=False)
    df3.to_excel(writer,startrow=len(df)+3,header=False,index=False)
    writer.save()

城市=['纽约','波士顿','洛杉矶','旧金山']
整体=['NYXRNSA.csv','BOXRNSA.csv','LXXRNSA.csv','SFXRNSA.csv']
高端=['NYXRHTNSA.csv','BOXRHTNSA.csv','LXXRHTNSA.csv','SFXRHTNSA.csv']
中端=['NYXRMTNSA.csv','BOXRMTNSA.csv','LXXRMTNSA.csv','SFXRMTNSA.csv']
低端=['NYXRLTNSA.csv','BOXRLTNSA.csv','LXXRLTNSA.csv','SFXRLTNSA.csv']
for a,b,c,d,e in zip(城市,整体,高端,中端,低端):
    市场(a,b,c,d,e)
芝加哥市场()
```
